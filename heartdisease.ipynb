```python
#Import all library needed
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense,BatchNormalization, Dropout
import tensorflow as tf
import matplotlib.pyplot as plt
import os
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

#confusion matrix visualization
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report
```

```python
### 1. Link notebook with google drive and access data from your personal Gdrive
from google.colab import drive
drive.mount("/content/gdrive")

### 2.Set the data path for dataset and model location (ex: model_loc = "/content/gdrive/My Drive/Dataset/")
dataset_dir = "/content/gdrive/My Drive/Heart Disease/Dataset/" #insert the path here
model_loc = "/content/gdrive/My Drive/Heart Disease/Model/" #insert the path here

print(os.listdir(dataset_dir))
data = pd.read_csv(dataset_dir+'heart.csv')
```

```python
### 3. Insert Exploratory data analysis (EDA) steps to analyze and investigate datasets.

data.head() # display first 5 row data
```

```python
### 4. What is the purpose of the code that sets a list of categorical variables
### in a dataset and then casts those variables to the object data type using the astype() function?
### Answer: helps in saving memory, making sure the data is represented correctly, and enabling special operations and analysis on categorical data

catagorialList = ['sex','cp','fbs','restecg','exang','ca','thal']
for item in catagorialList:
    data[item] = data[item].astype('object') #casting to object
```

```python
 ### 5. Create more data by categorical variable into indicator variables using 'get_dummies' function

data =  pd.get_dummies(data, drop_first=True)#complete your code here
```

```python
### 6. Explain line 3,4 and 5 and print the shape of x and y

y = data['target'].values # take target value
y = y.reshape(y.shape[0],1) # reshape y in 1 dimension
x = data.drop(['target'],axis=1) # drop target in x
```

```python
### 7. Create a simple dataset and demonstrate the normalization code on the simple dataset

# Create a simple dataset
data = pd.DataFrame({'A': [10, 20, 30], 'B': [100, 200, 300], 'C': [1000, 2000, 3000]})
print('Original dataset:')
print(data)

# Normalize data (range 0 - 1)
minx = np.min(data)
maxx = np.max(data)
data_norm = (data - minx) / (maxx - minx)
print('\nNormalized dataset:')
print(data_norm)
```

```python
### 8. Describe the heart dataset after implementing the min max normalization
#Normalize data (range 0 - 1)
minx = np.min(x)
maxx = np.max(x)
x = (x - minx) / (maxx - minx)
x.head()
#provide statistical information for each column in the normalized dataset
```

```python
### 9. Modify the code to split the dataset into train and test (train 70%, val 20% and test 10%).

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
# re-create train and validation set
x_train, x_val, y_train, y_val  = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
# train 70%, validation 20%, test 10%
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
```

```python
### 10. What is the purpose of each layer in the neural network created using the Sequential() function with 64, 32, and 1 neurons,
### respectively, and softmax and sigmoid activation functions?
### Answer: extract and transform the input data into a form that can be used to make predictions, with the final layer providing the output predictions

model = Sequential() #Allow us to create model layer by layer
model.add(Dense(64, input_dim=21, activation='softmax')) #Softmax turn number data into probabilities which sum to 1
model.add(Dense(32, activation='softmax'))
model.add(Dense(1, activation='sigmoid')) # produce probability value (number between 0 or 1)
model.summary()
```

```python
### 11. This code compiles a neural network model with a mean squared error loss function, the Adam optimizer with a learning rate of 0.01,
### and accuracy as a performance metric. What does each of these components mean, and how do they affect the model training and performance?

model.compile(loss='mse', ###MSE = computes the average of the squared differences between the predicted values and the actual target values
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999,epsilon=1e-07, amsgrad=False,name='Adam'), ### Adam optimizer = updating the model's parameters during training in order to minimize the loss function
              metrics=['acc'])
```

```python
# start the model training
output = []
early = EarlyStopping(monitor='val_acc', patience=400, mode='auto')
checkpoint = ModelCheckpoint(model_loc+"heart_disease_best_model.hdf5", monitor='val_acc', verbose=0, save_best_only=True, mode='auto', save_freq='epoch')
reduce_lr = ReduceLROnPlateau(monitor='val_acc', factor=0.01, patience=100, verbose=1, mode='auto', min_lr=0.001)
callbacks_list = [early]

output = model.fit(x_train, y_train,validation_data=(x_val,y_val), epochs=1000, batch_size=16, verbose=1, callbacks=callbacks_list)
```

```python
### 12. What does the plot generated by this code represent?

plt.plot(output.history['acc']) #represents the accuracy of the model on the training data for each epoch.
plt.plot(output.history['val_acc']) # represents the accuracy of the model on the validation data for each epoch
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
#plt.savefig('Accuracy.png',dpi=100) #to save the image
plt.show()
```

```python
### 13. Plot code for the model loss. You can refer to the plot code for model accuracy above.
plt.plot(output.history['loss'])
plt.plot(output.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
#plt.savefig('Loss.png',dpi=100) #to save the image
plt.show()
```

```python
### 14. What is the purpose of evaluating the model on the test dataset?
### Answer: performance evaluation

#model.load_weights(model_loc+"heart_disease_best_model.hdf5")
scores = model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
# print("\n%s: %.2f%%" % (model.metrics_names[0], scores[0]))
print("loss:", round(scores[0],2))
```

```python
#Display detailed prediction
pred = model.predict(x_test) #predicts the output labels
y = np.round(pred).astype("int16") #predicted values to the nearest integer and converts them to integers
idx = 0 # index variable for iteration
ps = 0
fl = 0 # Initialize variables to count the number of correct and incorrect predictions
for x in pred:
    if y_test[idx]==y[idx]:
        print("\033[30mNo:",idx+1,"Actual:",y_test[idx]," Predicted:",y[idx],"Result: \033[92mPass") #predicted label matches the actual label (y_test), it prints the prediction in green color indicating a pass
        ps = ps+1
    else:
        print("\033[30mNo:",idx+1,"Actual:",y_test[idx]," Predicted:",y[idx]," Result: \033[91mFail") #predicted label does not match the actual label, it prints the prediction in red color indicating a fail
        fl = fl+1
    idx = idx + 1
print("\033[30mRight Prediction :",ps, "Wrong Prediction :",fl)
```

```python
### 15. What is Confusion Matrix and why you need it? Explain TP, FP, FN, TN.
###True Positive (TP): These are the cases where the model predicted the positive class correctly, i.e., the actual value is positive, and the model also predicted it as positive.
###False Positive (FP): These are the cases where the model predicted the positive class incorrectly, i.e., the actual value is negative, but the model predicted it as positive.
###False Negative (FN): These are the cases where the model predicted the negative class incorrectly, i.e., the actual value is positive, but the model predicted it as negative.
###True Negative (TN): These are the cases where the model predicted the negative class correctly, i.e., the actual value is negative, and the model also predicted it as negative.

### 16. Explain the classification report produce.

y_pred = y
y_true = y_test

cm = confusion_matrix(y_true, y_pred, labels=[0,1])
#cm = confusion_matrix(y_true, y_pred, labels=labels.astype('int'))
f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidths=1.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
print()
print(classification_report(y_true, y_pred, labels=[0,1]))
```
